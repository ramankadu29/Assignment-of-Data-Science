```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
x
```

```python
name_of_company = ['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
```

```python
#Pie Plot 
plt.figure(figsize=(15,8))
sns.boxplot(x,orient="h")
plt.show()
```

```python
plt.pie(x,labels=name_of_company, autopct= '%.1f%%')
plt.show()
```

```python
describe = x.describe()
describe
```

```python
Variance = x.var()
Variance
```

```python
Q1 = np.quantile(x,0.25)
Q3 = np.quantile(x,0.75)
IQR = Q3 - Q1
med = np.median(x)
upper_bound = Q3+(1.5*IQR)
lower_bound = Q1-(1.5*IQR)

print('First Quantile=', Q1, 'Second Quantile=', med, 'Third Quantile=', Q3,
      'Inter-Quartile Range=', IQR, 'Upper Whisker=', upper_bound, 'Lower Whisker=', lower_bound)

Outliers = x[(x <= lower_bound) | (x >= upper_bound)]

print('The outlier in the boxplot:',Outliers)
```
